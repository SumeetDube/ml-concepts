```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

slope = (np.mean(x) * np.mean(y) - np.mean(x*y)) / (np.mean(x)**2 - np.mean(x**2))

intercept = np.mean(y) - slope * np.mean(x)

y_pred = slope * x + intercept

plt.scatter(x,y)
plt.plot(x,y_pred, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1, 2], [1, 4], [2, 2], [2, 4], [3, 3]])
y = np.array([6, 8, 8, 10, 12])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predicted values
y_pred = model.predict([[2, 5]])

# Model coefficients
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 6, 9, 12, 15])
```

```python
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(x_poly,y)
```

```python
y_pred = model.predict(poly.fit_transform([[6]]))
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
x_range = np.linspace(1,6,100).reshape(-1,1)
x_r_poly = poly.transform(x_range)
y_range_pred = model.predict(x_r_poly)

y_pred = model.predict(poly.transform([[6]]))

# Model coefficients
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='Data Points')
plt.plot(x_range, y_range_pred, color='blue', label='Polynomial Regression Curve')
plt.scatter(6, y_pred, color='green', label='Prediction for X=6', s=100, edgecolor='black')
plt.title('Polynomial Regression Model')
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
