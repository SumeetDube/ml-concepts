# Iris Dataset

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

data = load_iris()
x = data.data
y = data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
accuracy = model.score(x_test, y_test)
accuracy
```

```python
y_pred = model.predict(x_test)
```

```python
loss = sum(y_pred - y_test)**2/len(x_test)
loss
```

# Titanic Survival Prediction

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

data  = pd.read_csv('dataset/titanic/titanic.csv')
```

```python
x = data[['Pclass','Sex','Age','SibSp','Parch']]
y = data['Survived']
```

```python
x = pd.get_dummies(x,columns=['Sex'])
x
```

```python
x = x.fillna(x.mode().iloc[0])
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
accuracy = model.score(x_test,y_test)
accuracy
```

```python
y_pred = model.predict(x_test)
loss = sum((y_pred - y_test)**2)/len(y_test)
loss
```

# Boston Housing Price Prediciton

```python
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
data = fetch_california_housing()
```

```python
x = data.data
y = data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
mse = model.score(x_test,y_test)
```

```python
y_hat = model.predict(x_test)
mse = (1/len(x_test)) * sum(y_test - y_hat)**2
mse
```

```python
rss = model.score(x_test, y_test)
rss
```
