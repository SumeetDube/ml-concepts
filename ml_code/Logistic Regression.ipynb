```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import accuracy_score, classification_report
data = pd.read_csv('dataset/Melbourne/melb_data.csv')
y = data['Price']
```

```python
x = data[['Rooms','Suburb', 'Type', 'Bathroom','Car', 'Landsize','BuildingArea']]
x = pd.get_dummies(x, columns=['Suburb','Type'])
```

```python
x = x.fillna(x.mean())
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
for i in range(20):
    print(f'Ground Truth: {y_test.tolist()[i]} Predicted: {int(y_pred[i])}')
```

# Handwritten digit recongition

```python
import  matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
digits = load_digits()
x = digits.data
y = digits.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.imshow(confusion_matrix(y_test,y_pred))
```

```python
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
mse = sum(y_pred - y_test)**2/len(y_pred)
mse
```

# Stroke Prediciton

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

data = pd.read_csv('dataset/stroke/healthcare-dataset-stroke-data.csv')
```

```python
x = data.drop(['stroke','id'], axis=1)
y = data['stroke']
```

```python
x = pd.get_dummies(x, columns=['gender','ever_married', 'work_type','Residence_type','smoking_status'])
x = x.fillna(x.mean())
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
print('Classification Report:\n', classification_report(y_test, y_pred))
```
