```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
```

```python
data = load_iris()
```

```python
x, y = data.data, data.target
```

```python
tsne = TSNE(n_components=2, perplexity=30, max_iter=300)
x_embb = tsne.fit_transform(x)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(x_embb[:,0],x_embb[:,1], c=y,cmap='viridis')
plt.title('t-SNE visualization of Iris Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(scatter, label='Species')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X.shape
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
tsne = TSNE(n_components=2, perplexity=30, max_iter=300)
X_embedded = tsne.fit_transform(X_pca)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='viridis')
plt.title('t-SNE visualization of Iris Dataset after PCA')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(scatter, label='Species')
plt.show()
```
